```python
import scipy.stats as stats

def plot_histogram_and_qq(df):
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    sns.histplot(df, kde=True)
    plt.title('Histogram with KDE')
    
    plt.subplot(1, 2, 2)
    stats.probplot(df, dist="norm", plot=plt)
    plt.title('Q-Q Plot')
    
    plt.show()

plot_histogram_and_qq(df['price'])
```

```python
from scipy.stats import shapiro, kstest, normaltest

def check_normality(data, column_name):
    stat, p_shapiro = shapiro(data)
    stat, p_ks = kstest(data, 'norm', args=(data.mean(), data.std()))
    stat, p_dagostino = normaltest(data)
    
    print(f'Statistical Tests for {column_name}:')
    print(f'Shapiro-Wilk Test: p-value={p_shapiro}')
    print(f'Kolmogorov-Smirnov Test: p-value={p_ks}')
    print(f'D\'Agostino and Pearson\'s Test: p-value={p_dagostino}')
    
    alpha = 0.05
    if p_shapiro > alpha and p_ks > alpha and p_dagostino > alpha:
        print('Data is normally distributed (Fail to reject H0)')
    else:
        print('Data is not normally distributed (Reject H0)')
    print('\n')


check_normality(df['price'], 'price')
```
